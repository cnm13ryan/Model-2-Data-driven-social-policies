### We now want to calculate the mean of the result, confidence interval, standard deviation accordingly  

```python
def cal_stat_data(Dataframe, Range, exception=1):
    """
    Return the mean of the result, 95% CI (L), 95% CI (U), and SE
    """
    
    def matching(sex):
        if (sex == "Male"):
            return 0;
        if (sex == "Female"):
            return 1;
        else:
            return 2;
    
    def average(num, lst):
        for i in range(len(num)):
            tmp = lst[i] / num[i]
            lst[i] = tmp
        return lst;
    
    def condition(index):
        if (Dataframe['Indicator'][index] != Dataframe['Indicator'][Range[0]]):
            return 1;
    
    
    # [Male, Female, Persons]   
    tot_Result = [0, 0, 0];
    tot_U = [0, 0, 0];
    tot_L = [0, 0, 0];
    tot_SE = [0, 0, 0];
    num = [0, 0, 0];
    
    for i in range(Range[0], Range[1]):
        if condition(i) and exception:
            raise Exception(f"Sorry, indicator of {Range[0]} and {i} are not the same")
            
        sex = Dataframe['Sex'][i]
        k = matching(sex)
        num[k] += 1;
        tot_Result[k] += Dataframe['Result'][i];
        tot_U[k] += Dataframe['95% CI (U)'][i];
        tot_L[k] += Dataframe['95% CI (L)'][i];
        tot_SE[k] += Dataframe['SE'][i];
    
    a = average(num, tot_L)
    b = average(num, tot_Result)
    c = average(num, tot_U)
    d = average(num, tot_SE)
    return [a, b, c, d]
        
```

```python
def printResult(lst):
    
    def statement(tmp, num):
        sex = ["Male", "Female", "Person"]
        print(f"{sex[num]}: [95% CI (L):{round(tmp[0], 4)},  Result:{round(tmp[1], 4)},  95% CI (U):{round(tmp[2], 4)}, SE:{round(tmp[3], 4)}]")
        return;
    
    n = len(lst[0]);
    #print(n)
    # tmp = [CI (L), Result, CI (U), SE]
    tmp = [0, 0, 0, 0]
    
    for i in range(n):
        cnt = 0;
        for lst1 in lst:
            tmp[cnt] = lst1[i]
            cnt += 1
        statement(tmp, i)
    return

        
        
```

### Now we want to quantify self-rated mental health 

#### Let 
#### (1): Excellent / very good has a score of 2
#### (2): Good has a score of 1
#### (3): Fair/poor has a score of 0
 

### Since indicator (3) has a score of 0. Thus we can ignore it when calculating the mean score.

```python
def quantifyScore(lst2, lst1):
    
    def statement(tmp, num):
        sex = ["Male", "Female", "Person"]
        print(f"{sex[num]}: [95% CI (L):{round(tmp[0], 4)},  Result:{round(tmp[1], 4)},  95% CI (U):{round(tmp[2], 4)}, SE:{round(tmp[3], 4)}]")
        return;
    
    n = len(lst2[0]);
    #print(n)
    
    # tmp = [CI (L), Result, CI (U), SE]
    tmp = [0, 0, 0, 0]
    
    for i in range(n):
        j,k = 0, 0;
        for lst in lst2:
            tmp[j] += lst[i] * 1.5
            j += 1
            
        for lst in lst1:
            tmp[k] += lst[i] * 1
            k += 1   
        statement(tmp, i)
    return
```

### We now want to answer the first sub-question set as in above

#### Hypothesis test 1: Are there any mental health differences between males and females?

$H_{10} : \mu_{1m} = \mu_{1f}$

$H_{1A} : \mu_{1m} \neq \mu_{1f}$

$H_{10}$ suggest there is no difference between the self-rated mean mental health score of males $\mu_{1m}$ and females $\mu_{1f}$ from 2011 to 2020 in Australia

$H_{1A}$ suggest there is a significant difference between the self-rated mean mental health score of males $\mu_{1m}$ and females $\mu_{1f}$ from 2011 to 2020 in Australia


```python
# Mental Health
# Mean scores for males M1 and females M2
M1 = 120.04
M2 = 234.52 

# Standard Error for males SE1 and females SE2
SE1 = 7.23
SE2 = 14.49

# Sample size for males n1 and females n2
n1 = 10
n2 = 10
```

```python
def calTestStatic(M1, M2, SE1, SE2, n1, n2):
    
    numer = M1 - M2
    denom = math.sqrt(SE1**2 / n1 + SE2**2 / n2)
    return numer / denom

def calDegFreedom(n1, n2):
    return n1 + n2 - 2
```

```python
calTestStatic(M1, M2, SE1, SE2, n1, n2)
```

```python
calDegFreedom(n1, n2)
```

### We can easily see that the calculated t-value is greater than the critical t-value, such that we reject the null hypothesis with a confidence significance level of (?)
